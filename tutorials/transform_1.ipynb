In this tutorial, we will learn how to use pv-vision to 
crop a module image into single cells using our 'split' method. 


```python
img = cv.imread('../examples/transform_seg/module_imgs/example_2.jpg', 0)
raw_module = SplitModule(img, 6, 10, 3) # 6 rows && 10 columns && 3 busbar
plt.imshow(raw_module.image, 'gray')
```

```python
cells = raw_module.crop_cell(cellsize=250, hsplit=100, hthre=0.95, hinterval=50, vsplit=100, vthre=0.95, vinterval=300)
plt.imshow(cells[10], 'gray')
```
